## **Reflection on data models and management in ArcGIS Pro**

- **Write a 100-word paragraph which describes the relationship between**
  - a **`Geodatabase`**
  - a **`Feature Dataset`**
  - and **`Feature Classes`**
  
  Refer to **Page 9** of the assignment guide, and other resources online such as:
    - [Fundamentals of the geodatabase](https://pro.arcgis.com/en/pro-app/latest/help/data/geodatabases/overview/fundamentals-of-the-geodatabase.htm)
    - [Feature datasets in ArcGIS Pro](https://pro.arcgis.com/en/pro-app/latest/help/data/feature-datasets/feature-datasets-in-arcgis-pro.htm)



A geodatabase is a collection of various types of geographic datasets. It is a newer file storage system, which stores multiple related files that can be related by geography, type, etc. Inside a geodatabase there is something called a feature dataset. A feature dataset is a collection of spatial data, such as data layers and other geometric constructs. Multiple feature datasets can reside in a geodatabase. When you hold clusters of these feature datasets together you have layers called feature classes. To summarize the relationship of these three words, geodatabases hold feature datasets while feature datasets hold feature classes.
