## **Part 3: Reflection**

*In a brief paragraph, compare the two maps you created. Describe the visual and data model differences between the DEM datasets visualized in the topology map and the zonal statistics calculated and appended to the parcel dataset in the second map.*

<mark> Visually, the topography map shows more of what the actual area looks like as far as the land and the different elevations. The zonal map depicts the sections split up into different zones, as the name suggests, and shows less what the actual land looks like but aims to display the different zones and the mean of the slope within these zones. As far as data, the topography data existed on a gradient that was displayed in a gradient of colors as well. The zonal data split into nine classes and was  assigned colors for each class rather than being on a gradient scale. </mark>
